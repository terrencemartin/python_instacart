```python
# create groups
busy = [9,10,11,12,13,14,15,16,17]
```

```python
avg = [8,18,19]
```

```python
slow = [0,1,2,3,4,5,6,7,20,21,22,23]
```

```python
# create formula
result = []

for value in ord_prd_merge["order_hour_of_day"]:
  if value in busy:
    result.append("Most orders")
  elif value in slow:
    result.append("Least orders")
  else:
    result.append("Average orders")
```
